```python
from huggingface_hub import HfApi

# Define your Hugging Face space details
HF_USERNAME = "Prathamesh25" 
REPO_NAME = "unseen-aptitude-qa-dataset"  # Your target repository name

api = HfApi()
repo_id = f"{HF_USERNAME}/{REPO_NAME}"

# 1. Create the repository on the Hub (safely handles existing repos)
try:
    api.create_repo(repo_id=repo_id, repo_type="dataset", private=False)
    print(f"Created new repository: {repo_id}")
except Exception:
    print(f"Repository '{repo_id}' already exists, proceeding to upload.")

# 2. Push your generated JSONL file straight to the cloud
print("Uploading dataset file...")
api.upload_file(
    path_or_fileobj="metadata_aptitude_dataset.jsonl",
    path_in_repo="train.jsonl",  # Automatically readies it for Hugging Face Trainers
    repo_id=repo_id,
    repo_type="dataset"
)

print(f"🎉 Complete! View your dataset at: huggingface.co{repo_id}")
```
